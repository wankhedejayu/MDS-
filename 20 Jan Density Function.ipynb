```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
height =np.random.normal(165,7,1000)
```

```python
height
```

```python
plt.hist(height,bins =10)
plt.title("Height of Indians")
plt.xlabel("Height in CM")   
plt.ylabel("Distribution")
plt.show()
```

```python
dice= np.random.choice([1,2,3,4,5,6],1000)
```

```python
plt.hist(dice,bins =20)
plt.title("Dice")
plt.xlabel("Dice Number")   
plt.ylabel("Distribution")
plt.show()
```
