```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x= np.linspace(4,15,100)
y= np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,y,label="Sin Wave",linestyle="solid",linewidth=2,color="Blue")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Sin Wave")
plt.legend()
plt.grid(True)
plt.show()
```

### Bar Plot

```python
category = ["A", "B", "C", "D", "E"]
value=[33,54,42,51,29]
```

```python
plt.figure(figsize =(5,3))
plt.bar(category, value , color=["#234321","#234290","#234269","#234310" , "#234250"])
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Comparing Categories")
plt.legend()
plt.grid(True)
plt.show
```

### Scatter Plot

```python
x=np.random.rand(100)
y = np.random.rand(100)
color = np.random.rand(100)
size = 1000 * np.random.rand(100)
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x,y,c=color ,s=size , alpha=0.5 , cmap="plasma")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot")
plt.colorbar(label="Color Intensity")
plt.grid(True)
plt.show()
```

### Histogram

```python
dice= np.array([1,2,3,4,5,6])    #Dice
sample = np.random.choice(dice,4)# Rolling dice 4 time in a row and getting values
print("Output of 4 samples", sample)
print("Mean of above 4 samples", sample.mean())
```

```python
samples_1000 =[np.random.choice(dice,4).mean() for i in range(1000)]
samples_1000
```

```python
plt.figure(figsize=(5,3))
plt.hist(samples_1000, alpha=0.7 , edgecolor="black", bins=100, histtype="barstacked")
plt.show()
```

```python
data = np.random.randn(10000)
plt.figure(figsize=(5,3))
plt.hist(data, alpha=0.7, edgecolor="purple" , bins=100 , histtype="stepfilled")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.grid(True)
plt.show()
```

### Piechart

```python
sizes = np.random.randint(35,89,4)
labels=["East","West","North","South"]

plt.figure(figsize=(5,4))
plt.pie(sizes, labels=labels, colors=["gold","lightgreen","cyan","lightcoral"],
        autopct= "%1.1f%%", explode=[0, 0.1, 0.3, 0])
plt.title("Piechart")
plt.show()
```

### Boxplot

```python
data = np.random.normal(loc=0, scale=1, size=(100,))
plt.figure(figsize= (5,4))
plt.boxplot(data, patch_artist= True, vert= False)
plt.title("Boxplot")
plt.show()
```

### Polar Plot

```python
angles= np.linspace(0,2*np.pi, 100)
radii= np.random.rand(100)
```

```python
plt.figure(figsize=(5,4))
ax=plt.subplot(111, polar=True)
ax.plot(angles, radii, color="red",linewidth=0.2)
ax.fill(angles, radii , color="red", alpha=0.3)
plt.title("Polar Plot")
plt.show()
```

### Heatmap

```python
matrix= np.random.rand(5,5)
plt.figure(figsize=(5,4))
plt.imshow(matrix, cmap="plasma")
plt.title("Heatmap")
plt.colorbar(label= "Color Intensity")
plt.show()
```
